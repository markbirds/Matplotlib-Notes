### Figures and Axes
[Back to top](#top)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df1 = pd.read_csv('covid.csv')
df2 = pd.read_csv('sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
fig = plt.figure(facecolor='#ff9f36')
ax = fig.add_axes([0,0,1,1],fc='#ffe8cf')
dataframe = pd.crosstab(index=df2['Age Group'],columns=df2['Product Category'],values=df2['Order Quantity'],aggfunc='sum').reindex(['Youth (<25)','Young Adults (25-34)','Adults (35-64)','Seniors (64+)'])
xpos = np.arange(len(dataframe))

ax.set_xticks(xpos)
ax.set_xticklabels(dataframe.index)
ax.bar(xpos-0.2,dataframe.Accessories,width=0.2)
ax.bar(xpos,dataframe.Bikes,width=0.2)
ax.bar(xpos+0.2,dataframe.Clothing,width=0.2)
ax.legend(dataframe.columns)
ax.set_xlabel('Age Group')
ax.set_ylabel('Order Quantity')
ax.set_title('Order Quantity from different Products based on Age Group')
plt.show()
```

### Multiplot
[Back to top](#top)

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1.2,1.2])
ax3 = fig.add_axes([0.04,0.65,0.5,0.5])
ax2 = fig.add_axes([0.07,0.65,0.4,0.5])
ax3.set_xticks([])
ax3.set_yticks([])

ax1Data = df2[df2['Country']=='United States'].Revenue.resample('M').sum()
ax2Data = df2[df2['Country']=='United States'].groupby('Product Category').Revenue.sum()

ax1.stackplot(ax1Data.index,ax1Data)
ax2.pie(ax2Data,labels=ax2Data.index)

ax1.set_title('Monthly Revenue from Unites States')
plt.show()
```

### subplots and sublot2grid
[Back to top](#top)

```python
canada = df2[df2['Country']=='Canada'].Revenue.resample('M').sum()
australia = df2[df2['Country']=='Australia'].Revenue.resample('M').sum()
usa = df2[df2['Country']=='United States'].Revenue.resample('M').sum()
germany = df2[df2['Country']=='Germany'].Revenue.resample('M').sum()
france = df2[df2['Country']=='France'].Revenue.resample('M').sum()
uk = df2[df2['Country']=='United Kingdom'].Revenue.resample('M').sum()

fig,ax = plt.subplots(nrows=2,ncols=3,figsize=(13,8),sharey=True)
ax[0,0].stackplot(canada.index,canada,color='#a83232')
ax[0,1].stackplot(australia.index,australia,color='#1313bd')
ax[0,2].stackplot(usa.index,usa,color='#424bf5')
ax[1,0].stackplot(germany.index,germany,color='#f5ef42')
ax[1,1].stackplot(france.index,france,color='#f56942')
ax[1,2].stackplot(uk.index,uk,color='#f54242')

ax[0,0].set_title('Revenue from Canada')
ax[0,1].set_title('Revenue from Australia')
ax[0,2].set_title('Revenue from United States')
ax[1,0].set_title('Revenue from Germany')
ax[1,1].set_title('Revenue from France')
ax[1,2].set_title('Revenue from United Kingdom')

plt.show()
```

```python
ax1 = plt.subplot2grid((2,3),(0,0),rowspan=2,colspan=2)
ax2 = plt.subplot2grid((2,3),(0,2))
ax3 = plt.subplot2grid((2,3),(1,2))

plt.tight_layout()
dataframe1 = df1.groupby('WHO Region').Confirmed.sum().sort_values()
dataframe2 = df1[['Active','Recovered','Deaths']].sum()
dataframe3 = df1[['New cases','New recovered','New deaths']].sum()
ax1.barh(dataframe1.index,dataframe1/1000000)
ax1.set_title('Confirmed Cases from different Regions')
ax1.set_xlabel('Million')
ax2.pie(dataframe2,labels=dataframe2.index)
ax3.pie(dataframe3,labels=dataframe3.index)
plt.show()
```

### Formatting Axes
[Back to top](#top)

```python
male = df2[df2['Customer Gender']=='M'].Revenue.resample('M').count()
female = df2[df2['Customer Gender']=='F'].Revenue.resample('M').count()

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(df2.resample('M').sum().index,male,female,lw=3)
ax.legend(['Male','Female'])
ax.set_title('Customer Gender',pad=20)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.set_ylim(0,4000)
ax.set_xlim('2010','2018')
plt.show()
```

### More Scatter Plots
[Back to top](#top)

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

male = np.random.choice(np.arange(len(df2[df2['Customer Gender']=='M'])),size=100)
female = np.random.choice(np.arange(len(df2[df2['Customer Gender']=='F'])),size=100)

ax.scatter(df2.iloc[male]['Customer Age'],df2.iloc[male]['Order Quantity'])
ax.scatter(df2.iloc[female]['Customer Age'],df2.iloc[female]['Order Quantity'])
ax.set_title('Customer Age and Order Quantity')
ax.set_xlabel('Customer Age')
ax.set_ylabel('Order Quantity')
ax.legend(['Male','Female'])
plt.show()
```
