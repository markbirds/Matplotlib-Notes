### Working With Text
[Back to top](#top)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1.5,1.5])

revenue = df.Revenue.resample('M').sum()
maxPoint = (revenue.sort_values(ascending=False).index[0],revenue.sort_values(ascending=False)[0])

ax.stackplot(revenue.index,revenue)
ax.set_ylim(0,5000000)
ax.set_title('Monthly Revenue')
ax.plot(maxPoint[0],maxPoint[1],'o')
ax.annotate('Highest Monthly Revenue',maxPoint,
            xytext = ('2013-11-01',4000000),
            arrowprops = dict(facecolor='black',shrink = 0.05,width=0.1,headwidth=8),
            fontsize = 13)
ax.annotate('$ {:,}'.format(maxPoint[1]),maxPoint,xytext = ('2016-01-01',3950000),fontsize = 13)
ax.text('2011',4500000,'Average monthly revenue:',fontsize = 12)
ax.text('2011',4200000,'            {:,}'.format(round(revenue.mean())),fontsize = 12)
ax.plot(revenue.index,np.random.choice(list([round(revenue.mean())]),size=len(revenue)),'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xticks(np.arange(10))
ax.set_yticks(np.arange(10))
ax.set_title('Sample Formulas')
ax.text(0.6,8,'Formula for Binary Distribution', fontsize = 14)
ax.text(0.6, 7, r'$b(x;n,p) = \frac{n!}{x!(n-x)!}p^xq^{n-x}$', fontsize = 14)
ax.text(0.6,5,'Formula for Poisson Distribution', fontsize = 14)
ax.text(0.6, 4, r'$P(x) = \frac{(e)^{-lam}(lam)^x}{x!}$', fontsize = 14)
ax.text(0.6,1.5,
'P(x) is the poisson distribution\n' +
'e = 2.71828, a constant used in connection with natural logarithm\n' +
'lam =  mean or average in the distribution\n' +
'x is the specific value in which we are interested', fontsize = 11.5)
plt.show()
```
