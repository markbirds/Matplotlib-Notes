### Basic Plot
[Back to top](#top)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head(5)
```

```python
bikes = df[(df['Country'] == 'Canada') & (df['Product Category'] == 'Bikes')].Revenue.resample('M').sum()
```

```python
x = bikes.index
y = bikes
plt.plot_date(x,y,color='#fc4a03',fmt='-',lw=3)
plt.title('Monthly Revenue of Bikes from Canada')
plt.xlabel('Year')
plt.ylabel('Revenue in Dollars')
plt.grid()
plt.show()
```

```python
us = df[df['Country']=='United States'].Profit.resample('A').sum()
aus  = df[df['Country']=='Australia'].Profit.resample('A').sum()

plt.plot(us.index,us,lw=3)
plt.plot(aus.index,aus,lw=3)
plt.legend(['Unites States','Australia'])
plt.title('Sales Profit from US and Aus')
plt.xlabel('Year')
plt.ylabel('Profit in $')
plt.grid()
plt.show()
```

<a id='stackplot'></a>
### Stackplot
[Back to top](#top)

```python
accessories = df[df['Product Category']=='Accessories'].Revenue.resample('M').sum().reindex(index=pd.date_range('01/01/2011','07/31/2016',freq='M')).fillna(0)
clothing = df[df['Product Category']=='Clothing'].Revenue.resample('M').sum().reindex(index=pd.date_range('01/01/2011','07/31/2016',freq='M')).fillna(0)
bikes = df[df['Product Category']=='Bikes'].Revenue.resample('M').sum().reindex(index=pd.date_range('01/01/2011','07/31/2016',freq='M')).fillna(0)

plt.stackplot(bikes.index,bikes,accessories,clothing)
plt.xticks(['2011','2012','2013','2014','2015','2016'])
plt.legend(['Bikes','Clothing','Accesories'],loc=2)
plt.title('Revenue from different Products')
plt.show()
```

<a id='pie'></a>
### Pie Chart
[Back to top](#top)

```python
revCountries = df.groupby('Country').Profit.sum()

plt.pie(revCountries,labels=revCountries.index,autopct='%0.2f%%',explode=[0,0,0.2,0,0,0])
plt.title('Profit Percentage from different countries')
plt.axis('equal')
plt.show()
```

### Scatter Plot

```python
dataframe = pd.read_csv('sampleData2.csv')
```

```python
x = np.random.choice(dataframe.index, size=(383))
```

```python
age = dataframe.iloc[x]['Customer Age']
order = dataframe.iloc[x]['Order Quantity']
```

```python
plt.scatter(age,order)
plt.title('Relationship between Customer Age and Order Quantity')
plt.xlabel('Customer Age')
plt.ylabel('Order Quantity')
plt.show()
```
